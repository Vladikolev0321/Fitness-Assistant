# Strava activity api

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

Getting my activity data

```python
#print("Requesting Token...\n")
res = requests.post(STRAVA_AUTH_URL, data=STRAVA_DATA, verify=False)
access_token = res.json()['access_token']
#print("Access Token = {}\n".format(access_token))

header = {'Authorization': 'Bearer ' + access_token}
my_dataset = requests.get(STRAVA_ACTIVITIES_URL, headers=header).json()
```

```python
my_dataset
```

Printing my last activity type

```python
print(my_dataset[0]["name"])
```

```python
all_activities = json_normalize(my_dataset) # making the json into a pandas dataframe
```

### Exploring the data

```python
all_activities.head()
```

Extracting only needed columns

```python
cols = ['name', 'upload_id', 'type', 'distance', 'moving_time',   
         'average_speed', 'max_speed', 'start_date_local']
all_activities = all_activities[cols] # extracting only needed columns

all_activities.head()
```
